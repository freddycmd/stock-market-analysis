# 01 — Data Exploration

Initial data loading, cleaning, and exploratory data analysis of stock market data.

```python
import sys
sys.path.insert(0, '..')

from src.data_loader import StockDataLoader
from src.analysis import StockAnalyzer
from src.visualizer import StockVisualizer

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = (14, 6)
```

## Load Data

```python
loader = StockDataLoader(cache_dir='../data/raw')

# Fetch historical data for a stock
ticker = 'AAPL'
df = loader.fetch(ticker, start='2020-01-01')
df.head(10)
```

## Basic Info

```python
print(f'Shape: {df.shape}')
print(f'Date range: {df.index.min()} to {df.index.max()}')
print(f'\nMissing values:\n{df.isnull().sum()}')
print(f'\nSummary statistics:\n')
df.describe()
```

## Price Overview

```python
analyzer = StockAnalyzer(df)
analyzer.add_moving_averages([20, 50, 200])
analyzer.add_volume_sma()

viz = StockVisualizer(analyzer.get_data(), ticker=ticker)
viz.fig_dir = '../reports/figures'

viz.plot_price(save=False)
plt.show()
```

## Volume Analysis

```python
viz.plot_volume(save=False)
plt.show()
```

## Returns Distribution

```python
viz.plot_returns_distribution(save=False)
plt.show()
```

## Summary Statistics

```python
stats = analyzer.summary_stats()
for k, v in stats.items():
    print(f'{k:>25s}: {v:>10.4f}')
```
