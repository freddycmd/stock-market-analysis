# 02 — Technical Analysis

Technical indicators: RSI, MACD, Bollinger Bands, and trading signals.

```python
import sys
sys.path.insert(0, '..')

from src.data_loader import StockDataLoader
from src.analysis import StockAnalyzer
from src.visualizer import StockVisualizer

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (14, 6)
```

```python
loader = StockDataLoader(cache_dir='../data/raw')
ticker = 'AAPL'
df = loader.fetch(ticker, start='2020-01-01')

analyzer = StockAnalyzer(df)
analyzer.add_moving_averages([20, 50, 200])
analyzer.add_ema([12, 26])
analyzer.add_rsi()
analyzer.add_macd()
analyzer.add_bollinger_bands()

viz = StockVisualizer(analyzer.get_data(), ticker=ticker)
viz.fig_dir = '../reports/figures'
```

## Moving Averages

```python
viz.plot_price_with_ma()
plt.show()
```

## RSI (Relative Strength Index)

```python
viz.plot_rsi(save=False)
plt.show()
```

## MACD

```python
viz.plot_macd(save=False)
plt.show()
```

## Bollinger Bands

```python
data = analyzer.get_data()
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(data.index, data['Close'], label='Close', linewidth=1.5)
ax.plot(data.index, data['BB_Upper'], '--', label='Upper Band', color='#F44336', alpha=0.7)
ax.plot(data.index, data['BB_Middle'], '--', label='SMA 20', color='#FF9800', alpha=0.7)
ax.plot(data.index, data['BB_Lower'], '--', label='Lower Band', color='#4CAF50', alpha=0.7)
ax.fill_between(data.index, data['BB_Upper'], data['BB_Lower'], alpha=0.1, color='#9C27B0')
ax.set_title(f'{ticker} — Bollinger Bands', fontsize=14, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```
