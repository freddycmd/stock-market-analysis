# 04 — Interactive Dashboards

Interactive Plotly dashboards: candlestick charts, technical dashboard, returns heatmap, volume profile, and multi-stock comparison.

```python
import sys
sys.path.insert(0, '..')

from src.data_loader import StockDataLoader
from src.analysis import StockAnalyzer
from src.dashboard import StockDashboard
from src.charts import AdvancedCharts

import matplotlib.pyplot as plt

%matplotlib inline
```

## Load and Prepare Data

```python
loader = StockDataLoader(cache_dir='../data/raw')
ticker = 'AAPL'
df = loader.fetch(ticker, start='2020-01-01')

# Add all indicators
analyzer = StockAnalyzer(df)
analyzer.add_moving_averages([20, 50, 200])
analyzer.add_rsi()
analyzer.add_macd()
analyzer.add_bollinger_bands()
analyzer.add_volume_sma()

enriched = analyzer.get_data()
print(f'Loaded {ticker}: {len(enriched)} rows, {len(enriched.columns)} columns')
```

## Candlestick Chart

```python
dash = StockDashboard(enriched, ticker=ticker)

# Last 6 months of candlesticks
fig = dash.candlestick_chart(last_n_days=120)
fig.show()
```

## Full Technical Dashboard
Price + Bollinger Bands + Volume + RSI + MACD — all interactive.

```python
fig = dash.technical_dashboard()
fig.show()

# Save as interactive HTML
dash.save_html(fig, f'{ticker}_technical_dashboard.html')
```

## Monthly Returns Heatmap

```python
fig = dash.returns_heatmap()
fig.show()
```

## Volume Profile

```python
fig = dash.volume_profile(bins=60)
fig.show()
```

## Multi-Stock Comparison

```python
# Load comparison stocks
compare_tickers = ['MSFT', 'GOOGL', 'AMZN', 'TSLA']
others = {}
for t in compare_tickers:
    others[t] = loader.fetch(t, start='2020-01-01')

fig = dash.comparison_chart(others)
fig.show()
```
